# Heirarchical Clustering

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
crime=pd.read_csv(r'C:\Users\HP\Downloads\Excel Downloaded files\crime_data.csv')
crime
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
df_norm=norm_func(crime.iloc[:,1:])
```

```python
df_norm
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='average'))
```

```python
# create clusters
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc=hc.fit_predict(df_norm)
```

```python
y_hc
```

```python
crime['h_clusterid']=hc.labels_
crime
```

```python
crime.groupby('h_clusterid').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime['h_clusterid'],crime['Assault'], c=hc.labels_) 
```

# K means Clustering

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
crime1=pd.read_csv(r'C:\Users\HP\Downloads\Excel Downloaded files\crime_data.csv')
crime1
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
df_norm=norm_func(crime1.iloc[:,1:])
```

```python
# Elbow curv
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above  plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime1['Clust']=x
```

```python
crime1.iloc[:,1:5].groupby(crime1.Clust).mean()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime1['Clust'],crime['Assault'], c=x) 
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime2=pd.read_csv(r'C:\Users\HP\Downloads\Excel Downloaded files\crime_data.csv')
crime2
```

```python
crime2.info()
```

```python
crime2.drop(['Unnamed: 0'],axis=1,inplace=True)
crime2
```

```python
# Normalize heterogenous numerical data using standard scalar fit transform to dataset
crime2_norm=StandardScaler().fit_transform(crime2)
crime2_norm
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(crime2_norm)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
# Adding clusters to dataset
crime2['clusters']=dbscan.labels_
crime2
```

```python
crime2.groupby('clusters').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime2['clusters'],crime['UrbanPop'], c=dbscan.labels_) 
```
